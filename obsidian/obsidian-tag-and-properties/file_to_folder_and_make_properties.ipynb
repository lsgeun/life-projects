현재 디렉토리에 있는 파일과 동일한 이름을 가진 폴더를 만든 후 그곳에 파일을 각각 옮긴다.  
그 다음 만든 폴더와 동일한 이름을 가진 md 파일을 만들어서 옮겨진 파일에 대한 링크를 걸고, 이 파일과 관련된 properties를 만든다.

```python
import os
import re
path = "/Users/isgeun/Library/Mobile Documents/iCloud~md~obsidian/Documents/노트/7 참고 자료/책/컴퓨터/Python"
ref_type = "책"
resource = ["Python", "컴퓨터구조"]

file_names = os.listdir(path); file_names = [_ for _ in file_names if re.search(r".+\..+", _) != None]
file_folders = [re.search(r"(.+)\..+", _).group(1) for _ in file_names]

for _ in file_names:
    file_folder = re.search(r"(.+)\..+", _).group(1)
    
    os.system(f"mkdir '{path}/{file_folder}'")
    os.system(f"mv '{path}/{_}' '{path}/{file_folder}/{_}'")

for i, _ in enumerate(file_folders):
    os.system(f"touch '{path}/{_}/{_}.md'")
    
    with open(f'{path}/{_}/{_}.md', "w") as f:
        s = "---\n"
        s += f"참고_자료_유형: {ref_type}\n"
        s += "관심사:\n"
        s += f"  - {resource[0]}\n"
        # s += f"  - {resource[1]}\n"
        s += "---\n"
        s += "\n"
        s += "# 참고 자료\n"
        s += f"[{_}]({file_names[i].replace(" ", "%20")})\n"
        
        f.write(s)
```
